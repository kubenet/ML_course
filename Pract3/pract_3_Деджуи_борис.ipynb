## 📚 Теоретическая справка

### Линейная регрессия
Модель предсказывает целевую переменную как взвешенную сумму признаков:

$$y = w_1x_1 + w_2x_2 + ... + w_nx_n + b$$

### Функция потерь MSE
$$MSE = \frac{1}{n} \sum_{i=1}^{n} (y_i - \hat{y}_i)^2$$

### Градиентный спуск
$$w_{new} = w_{old} - \eta \cdot \nabla MSE$$

где $\eta$ — скорость обучения (learning rate)

### Задание 2: Реализация градиентного спуска с нуля

````python
class GradientDescentLinearRegression:
    """
    Реализация линейной регрессии с градиентным спуском с нуля
    """

    def __init__(self, learning_rate=0.01, n_iterations=1000):
        self.learning_rate = learning_rate
        self.n_iterations = n_iterations
        self.weights = None
        self.bias = None
        self.loss_history = []

    def fit(self, X, y):
        """
        Обучение модели с помощью градиентного спуска
        """
        n_samples, n_features = X.shape

        # Инициализация параметров
        self.weights = np.zeros(n_features)
        self.bias = 0

        # Градиентный спуск
        for i in range(self.n_iterations):
            # Предсказание
            y_pred = np.dot(X, self.weights) + self.bias

            # Вычисление градиентов
            dw = (2/n_samples) * np.dot(X.T, (y_pred - y))
            db = (2/n_samples) * np.sum(y_pred - y)

            # Обновление параметров
            self.weights -= self.learning_rate * dw
            self.bias -= self.learning_rate * db

            # Сохранение истории потерь
            mse = np.mean((y_pred - y) ** 2)
            self.loss_history.append(mse)

            # Вывод прогресса каждые 100 итераций
            if i % 100 == 0:
                print(f"Итерация {i}: MSE = {mse:.4f}")

        return self

    def predict(self, X):
        """
        Предсказание целевой переменной
        """
        return np.dot(X, self.weights) + self.bias

    def score(self, X, y):
        """
        Вычисление R² score
        """
        y_pred = self.predict(X)
        ss_res = np.sum((y - y_pred) ** 2)
        ss_tot = np.sum((y - np.mean(y)) ** 2)
        return 1 - (ss_res / ss_tot)
````

````python
# Обучаем нашу модель
print("Обучение модели с градиентным спуском...")
start_time = time.time()

gd_model = GradientDescentLinearRegression(learning_rate=0.01, n_iterations=1000)
gd_model.fit(X_train_scaled, y_train)

gd_time = time.time() - start_time
print(f"\nВремя обучения: {gd_time:.4f} секунд")
print(f"Веса модели: {gd_model.weights.round(4)}")
print(f"Смещение (bias): {gd_model.bias:.4f}")
````

````python
# Интерактивная визуализация влияния learning rate
from ipywidgets import interact, FloatSlider, IntSlider

def plot_gradient_descent_interactive(learning_rate=0.01, n_iterations=100):
    """
    Интерактивная демонстрация влияния параметров на сходимость
    """
    model = GradientDescentLinearRegression(
        learning_rate=learning_rate,
        n_iterations=n_iterations
    )
    model.fit(X_train_scaled[:100], y_train[:100])  # Используем подвыборку для скорости

    plt.figure(figsize=(12, 4))

    plt.subplot(1, 2, 1)
    plt.plot(model.loss_history, linewidth=2, color='blue')
    plt.xlabel('Итерация', fontsize=12)
    plt.ylabel('MSE', fontsize=12)
    plt.title(f'LR={learning_rate}, Iter={n_iterations}', fontsize=14)
    plt.grid(True, alpha=0.3)
    plt.yscale('log')

    plt.subplot(1, 2, 2)
    plt.bar(range(len(model.weights)), model.weights, color='green', alpha=0.7)
    plt.xlabel('Признак', fontsize=12)
    plt.ylabel('Вес', fontsize=12)
    plt.title('Веса модели', fontsize=14)
    plt.grid(True, alpha=0.3, axis='y')

    plt.tight_layout()
    plt.show()

    print(f"Финальное MSE: {model.loss_history[-1]:.4f}")
    print(f"R² score: {model.score(X_train_scaled[:100], y_train[:100]):.4f}")

# Создаём интерактивный виджет
interact(plot_gradient_descent_interactive,
         learning_rate=FloatSlider(value=0.01, min=0.001, max=0.1, step=0.001),
         n_iterations=IntSlider(value=100, min=10, max=500, step=10));
````

### Задание 4: Сравнение трёх подходов

````python
# 1. Градиентный спуск (наша реализация)
start_time = time.time()
gd_model = GradientDescentLinearRegression(learning_rate=0.01, n_iterations=1000)
gd_model.fit(X_train_scaled, y_train)
gd_pred = gd_model.predict(X_test_scaled)
gd_time = time.time() - start_time

# 2. Нормальное уравнение (аналитическое решение)
start_time = time.time()
X_train_bias = np.column_stack([np.ones(X_train_scaled.shape[0]), X_train_scaled])
# w = (X^T X)^(-1) X^T y
normal_weights = np.linalg.inv(X_train_bias.T.dot(X_train_bias)).dot(X_train_bias.T).dot(y_train)
normal_pred = np.column_stack([np.ones(X_test_scaled.shape[0]), X_test_scaled]).dot(normal_weights)
normal_time = time.time() - start_time

# 3. LinearRegression из sklearn
start_time = time.time()
sklearn_model = LinearRegression()
sklearn_model.fit(X_train_scaled, y_train)
sklearn_pred = sklearn_model.predict(X_test_scaled)
sklearn_time = time.time() - start_time

# Сравниваем результаты
print("=" * 70)
print("СРАВНЕНИЕ ТРЁХ ПОДХОДОВ")
print("=" * 70)

methods = ['Градиентный спуск', 'Нормальное уравнение', 'sklearn LinearRegression']
times = [gd_time, normal_time, sklearn_time]
mse_scores = [
    mean_squared_error(y_test, gd_pred),
    mean_squared_error(y_test, normal_pred),
    mean_squared_error(y_test, sklearn_pred)
]
r2_scores = [
    r2_score(y_test, gd_pred),
    r2_score(y_test, normal_pred),
    r2_score(y_test, sklearn_pred)
]

comparison_df = pd.DataFrame({
    'Метод': methods,
    'Время (сек)': [f'{t:.4f}' for t in times],
    'MSE': [f'{m:.4f}' for m in mse_scores],
    'R²': [f'{r:.4f}' for r in r2_scores]
})

print(comparison_df.to_string(index=False))
print("=" * 70)
````

##  Технические требования к генерации данных

```python
def generate_dataset(n_samples=500, n_features=2, noise_std=0.5,
                     outlier_fraction=0.05, outlier_multiplier=3,
                     random_seed=None):
    """
    Генерация синтетического датасета для линейной регрессии
    
    Parameters:
    -----------
    n_samples : int - количество наблюдений
    n_features : int - количество признаков
    noise_std : float - стандартное отклонение шума
    outlier_fraction : float - доля выбросов (0.0-0.2)
    outlier_multiplier : float - множитель для выбросов
    random_seed : int - seed для воспроизводимости
    """
    if random_seed is not None:
        np.random.seed(random_seed)
    
    # Генерация признаков
    X = np.random.randn(n_samples, n_features)
    
    # Истинные коэффициенты
    true_coef = np.random.uniform(1, 5, n_features)
    true_intercept = np.random.uniform(0, 10)
    
    # Генерация целевой переменной
    y = X @ true_coef + true_intercept + np.random.randn(n_samples) * noise_std
    
    # Добавление выбросов
    n_outliers = int(n_samples * outlier_fraction)
    outlier_idx = np.random.choice(n_samples, n_outliers, replace=False)
    y[outlier_idx] += np.random.choice([-1, 1], n_outliers) * \
                      np.random.uniform(3, 10, n_outliers) * outlier_multiplier
    
    return X, y, true_coef, true_intercept
```

#Этап 1: Генерация и подготовка данных


````python


import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split

# ============================================================
# STEP 1 — DATA GENERATION AND PREPARATION
# ============================================================

# Student variant parameters
student_id = 7
noise_std = 0.8
outlier_fraction = 0.08
outlier_multiplier = 4
features = ["experience", "education"]  # from variant 2


# ------------------------------------------------------------
# 1.1 Function to generate synthetic dataset (from подсказка 1)
# ------------------------------------------------------------
def generate_dataset(n_samples, n_features, noise_std, outlier_fraction,
                     outlier_multiplier, random_seed):
    """
    Generates a synthetic linear regression dataset with noise and outliers.
    All steps are explained for beginners.

    Parameters:
        n_samples (int): number of data points
        n_features (int): number of input features
        noise_std (float): standard deviation of Gaussian noise
        outlier_fraction (float): percentage of outliers
        outlier_multiplier (float): how strong outliers are
        random_seed (int): seed for reproducibility

    Returns:
        X (array): feature matrix
        y (array): target values
        true_w (array): true weights used to generate data
        true_b (float): true bias
    """

    np.random.seed(random_seed)

    # Generate features
    experience = np.random.uniform(0, 20, size=n_samples)
    education = np.random.randint(1, 5, size=n_samples)

    X = np.column_stack([experience, education])

    # True underlying weights (unknown to the model)
    true_w = np.array([2000, 5000])  # salary increases with experience & education
    true_b = 20000  # base salary

    # Generate clean target values
    y_clean = X @ true_w + true_b

    # Add Gaussian noise
    noise = np.random.normal(0, noise_std * 1000, size=n_samples)
    y = y_clean + noise

    # Add outliers
    n_outliers = int(n_samples * outlier_fraction)
    outlier_idx = np.random.choice(n_samples, n_outliers, replace=False)

    y[outlier_idx] += outlier_multiplier * np.random.choice(
        [-1, 1], size=n_outliers
    ) * np.abs(np.random.normal(0, 10000, size=n_outliers))

    return X, y, true_w, true_b, outlier_idx


# ------------------------------------------------------------
# 1.2 Generate dataset using student variant parameters
# ------------------------------------------------------------
X, y, true_w, true_b, outlier_idx = generate_dataset(
    n_samples=500,
    n_features=len(features),
    noise_std=noise_std,
    outlier_fraction=outlier_fraction,
    outlier_multiplier=outlier_multiplier,
    random_seed=student_id
)

# Create DataFrame for EDA
df = pd.DataFrame(X, columns=features)
df["salary"] = y

print("First 5 rows of the generated dataset:")
print(df.head())


# ------------------------------------------------------------
# 1.3 Train-test split (70/30)
# ------------------------------------------------------------
X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.30, random_state=student_id
)

print("\nTrain shape:", X_train.shape)
print("Test shape:", X_test.shape)


# ------------------------------------------------------------
# 1.4 Feature scaling
# ------------------------------------------------------------
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)

print("\nFeature scaling completed.")
print("Scaled mean:", X_train_scaled.mean(axis=0).round(2))
print("Scaled std:", X_train_scaled.std(axis=0).round(2))
````

#Этап 3: Обучение моделей

````python
# ============================================================
# STEP 3 — MODEL TRAINING
# ============================================================

from sklearn.linear_model import LinearRegression, HuberRegressor, SGDRegressor

# ------------------------------------------------------------
# 3.1 Train Linear Regression (minimizes MSE)
# ------------------------------------------------------------
model_linear = LinearRegression()
model_linear.fit(X_train_scaled, y_train)

print("Linear Regression trained successfully.")
print("Coefficients:", model_linear.coef_)
print("Intercept:", model_linear.intercept_)


# ------------------------------------------------------------
# 3.2 Train Huber Regressor (epsilon = 1.5 from variant)
# ------------------------------------------------------------
model_huber = HuberRegressor(
    epsilon=1.5,      # from variant 2
    max_iter=1000
)
model_huber.fit(X_train_scaled, y_train)

print("\nHuber Regressor trained successfully.")
print("True weights:", true_w)
print("Huber weights:", model_huber.coef_)
print("Huber intercept:", model_huber.intercept_)


# ------------------------------------------------------------
# 3.3 Train SGDRegressor with Huber loss
# ------------------------------------------------------------
model_sgd = SGDRegressor(
    loss="huber",         # from variant 2
    learning_rate="constant",
    eta0=0.005,           # learning rate from variant 2
    max_iter=1500,        # iterations from variant 2
    random_state=student_id
)

model_sgd.fit(X_train_scaled, y_train)

print("\nSGD Regressor (Huber loss) trained successfully.")
print("SGD weights:", model_sgd.coef_)
print("SGD intercept:", model_sgd.intercept_)
````

#Этап 4: Оценка качества


````python
# ============================================================
# STEP 4 — MODEL EVALUATION
# ============================================================

from sklearn.metrics import mean_squared_error, mean_absolute_error, r2_score

# ------------------------------------------------------------
# 4.1 Function to evaluate models
# ------------------------------------------------------------
def evaluate_model(model, X_test, y_test):
    """
    Computes evaluation metrics for a regression model.
    Returns a dictionary with MSE, RMSE, MAE, and R2.
    """
    y_pred = model.predict(X_test)
    mse = mean_squared_error(y_test, y_pred)
    rmse = np.sqrt(mse)
    mae = mean_absolute_error(y_test, y_pred)
    r2 = r2_score(y_test, y_pred)

    return {
        "y_pred": y_pred,
        "MSE": mse,
        "RMSE": rmse,
        "MAE": mae,
        "R2": r2
    }


# Evaluate all three models
results_linear = evaluate_model(model_linear, X_test_scaled, y_test)
results_huber = evaluate_model(model_huber, X_test_scaled, y_test)
results_sgd = evaluate_model(model_sgd, X_test_scaled, y_test)

print("=== Evaluation Results ===")
print("\nLinear Regression:", results_linear)
print("\nHuber Regressor:", results_huber)
print("\nSGD Regressor (Huber):", results_sgd)


# ------------------------------------------------------------
# 4.2 Function to compare models (подсказка 3)
# ------------------------------------------------------------
def compare_models(models_dict, X_test, y_test):
    results = []
    for name, model in models_dict.items():
        y_pred = model.predict(X_test)
        results.append({
            "Model": name,
            "MSE": mean_squared_error(y_test, y_pred),
            "MAE": mean_absolute_error(y_test, y_pred),
            "R2": r2_score(y_test, y_pred)
        })
    return pd.DataFrame(results)


models = {
    "LinearRegression": model_linear,
    "HuberRegressor": model_huber,
    "SGDRegressor (Huber)": model_sgd
}

comparison_df = compare_models(models, X_test_scaled, y_test)
print("\n=== Model Comparison Table ===")
print(comparison_df)


# ------------------------------------------------------------
# 4.3 Residual plot function (подсказка 4)
# ------------------------------------------------------------
def plot_residuals(y_true, y_pred, title="Residual Plot"):
    residuals = y_true - y_pred
    plt.figure(figsize=(7, 5))
    plt.scatter(y_pred, residuals, alpha=0.6, edgecolor="k")
    plt.axhline(0, color="red", linestyle="--")
    plt.xlabel("Predicted values")
    plt.ylabel("Residuals")
    plt.title(title)
    plt.grid(alpha=0.3)
    plt.show()


# ------------------------------------------------------------
# 4.4 Actual vs Predicted (Linear Regression)
# ------------------------------------------------------------
plt.figure(figsize=(7, 5))
plt.scatter(y_test, results_linear["y_pred"], alpha=0.6, edgecolor="k")
min_val = min(y_test.min(), results_linear["y_pred"].min())
max_val = max(y_test.max(), results_linear["y_pred"].max())
plt.plot([min_val, max_val], [min_val, max_val], "r--", label="Ideal line")
plt.xlabel("Actual salary")
plt.ylabel("Predicted salary")
plt.title("Actual vs Predicted (Linear Regression)")
plt.legend()
plt.grid(True)
plt.show()


# ------------------------------------------------------------
# 4.5 Residual plot for Linear Regression
# ------------------------------------------------------------
plot_residuals(y_test, results_linear["y_pred"], title="Residuals (Linear Regression)")


# ------------------------------------------------------------
# 4.6 Comparison of predictions of all three models
# ------------------------------------------------------------
plt.figure(figsize=(8, 6))

plt.scatter(y_test, results_linear["y_pred"], alpha=0.5, label="LinearRegression")
plt.scatter(y_test, results_huber["y_pred"], alpha=0.5, label="HuberRegressor")
plt.scatter(y_test, results_sgd["y_pred"], alpha=0.5, label="SGDRegressor (Huber)")

plt.plot([min_val, max_val], [min_val, max_val], "k--", label="Ideal line")

plt.xlabel("Actual salary")
plt.ylabel("Predicted salary")
plt.title("Comparison of Predictions (3 Models)")
plt.legend()
plt.grid(True)
plt.show()
````

````python
# ============================================================
# STEP 5 — ROBUSTNESS ANALYSIS
# ============================================================

# In this step, we analyze how sensitive each model is to outliers.
# We will compare model performance on:
# 1) full dataset (with outliers)
# 2) cleaned dataset (outliers removed)


# ------------------------------------------------------------
# 5.1 Remove outliers from the dataset
# ------------------------------------------------------------
df_clean = df.drop(index=outlier_idx).reset_index(drop=True)

X_clean = df_clean[["experience", "education"]].values
y_clean = df_clean["salary"].values

# Train-test split again
X_train_c, X_test_c, y_train_c, y_test_c = train_test_split(
    X_clean, y_clean, test_size=0.30, random_state=student_id
)

# Scale again
scaler_clean = StandardScaler()
X_train_c_scaled = scaler_clean.fit_transform(X_train_c)
X_test_c_scaled = scaler_clean.transform(X_test_c)


# ------------------------------------------------------------
# 5.2 Retrain all models on CLEAN data
# ------------------------------------------------------------
model_linear_clean = LinearRegression()
model_linear_clean.fit(X_train_c_scaled, y_train_c)

model_huber_clean = HuberRegressor(epsilon=1.5, max_iter=1000)
model_huber_clean.fit(X_train_c_scaled, y_train_c)

model_sgd_clean = SGDRegressor(
    loss="huber",
    learning_rate="constant",
    eta0=0.005,
    max_iter=1500,
    random_state=student_id
)
model_sgd_clean.fit(X_train_c_scaled, y_train_c)


# ------------------------------------------------------------
# 5.3 Evaluate models on CLEAN data
# ------------------------------------------------------------
results_linear_clean = evaluate_model(model_linear_clean, X_test_c_scaled, y_test_c)
results_huber_clean = evaluate_model(model_huber_clean, X_test_c_scaled, y_test_c)
results_sgd_clean = evaluate_model(model_sgd_clean, X_test_c_scaled, y_test_c)

print("\n=== CLEAN DATA RESULTS ===")
print("Linear Regression:", results_linear_clean)
print("Huber Regressor:", results_huber_clean)
print("SGD Regressor:", results_sgd_clean)


# ------------------------------------------------------------
# 5.4 Compare performance BEFORE and AFTER removing outliers
# ------------------------------------------------------------
comparison_robustness = pd.DataFrame({
    "Model": ["Linear", "Huber", "SGD"],
    "MSE (with outliers)": [
        results_linear["MSE"],
        results_huber["MSE"],
        results_sgd["MSE"]
    ],
    "MSE (clean data)": [
        results_linear_clean["MSE"],
        results_huber_clean["MSE"],
        results_sgd_clean["MSE"]
    ]
})

print("\n=== ROBUSTNESS COMPARISON TABLE ===")
print(comparison_robustness)


# ------------------------------------------------------------
# 5.5 Plot comparison of MSE before/after cleaning
# ------------------------------------------------------------
plt.figure(figsize=(8, 5))
x = np.arange(3)

plt.bar(x - 0.15, comparison_robustness["MSE (with outliers)"], width=0.3, label="With outliers")
plt.bar(x + 0.15, comparison_robustness["MSE (clean data)"], width=0.3, label="Clean data")

plt.xticks(x, ["Linear", "Huber", "SGD"])
plt.ylabel("MSE")
plt.title("Model Robustness: MSE Before/After Removing Outliers")
plt.legend()
plt.grid(alpha=0.3)
plt.show()
````
